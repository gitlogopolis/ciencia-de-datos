```python
df = pd.read_csv("2008-3.csv")
```

```python
len(dfduplicate) #muestra numero de filas
```

```python
dfduplicate = dfduplicate.sample(frac = 1)  #Vamos a reordernarlo
```

```python
dfclean = dfduplicate.drop_duplicates()  #Con esta instrucción se eliminan todos los casos duplicados
```
