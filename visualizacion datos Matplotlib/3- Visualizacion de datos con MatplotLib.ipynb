REALIZAR Y PERSONALIZAR UN GRAFICO DE BURBUJAS

```python
import seaborn as sns
df = pd.read_csv('base_datos_2008.csv', nrows=1000000)
df = df.sample(frac=1).head(100)
```

```python
plt.scatter(x=df.DayofMonth, y=df.ArrDelay, s=df.Distance)
```

```python
plt.scatter(x=df.DayofMonth, y=df.ArrDelay, s=df.Distance, alpha=0.3, c=df.DayOfWeek.isin([6,7])) #Diagrama,ejex, ejey,diametro burbuja, trasparencia burbuja, diferencia de color
plt.title("Retrasos en EEUU") #Titulo
plt.ylabel("Retraso al llegar") #Titulo en el eje y
plt.xlabel("Dia del Mes") #Titulo en el eje x
plt.ylim([0,150]) #Limite de vision en el eje y
plt.xticks([0,15,30]) #Valores del eje x
plt.text(x=28, y=120, s="Mi vuelo") #Texto en una coordenada
plt.show()
```

ETIQUETAS Y LEYENDAS EN MATPLOTLIB

```python
data = np.unique(df.DayOfWeek, return_counts=True) #Tabla Resumen
labs = ["Lun", "Mar","Mie","Jue","Vie","Sab","Dom"]
data
```

```python
plt.pie(x= data[1], #Porcentaje
        labels= labs, #Etiquetas
        radius=0.7, #Tamao de radio
        colors= sns.color_palette("hls",7), #Paleta de colores
        #colors=["Red","Green","Orange","Blue","Gray","Pink","Black"],
        explode = (0,0,0,0,0,0,0.1), #Selecciona el septimo valor
        startangle=90, 
        autopct= "%1.1f%%", #Muestra porcentaje del grafico
        )
#plt.legend(loc="upper left",labels=labs)
plt.show()
```

```python
plt = sns.barplot(x=labs, y=data[1]) #Generia un grafico de barras
plt.set(xlabel = "Dia de la semana", ylabel="Numero de vuelos") #Establece los labels
```
